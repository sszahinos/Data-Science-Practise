```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("./source/Bitcoin Historical Data.csv")
data
```

```python
cols = ["Price", "Open","High", "Low"]
for col in cols:
    data[col] = data[col].str.replace(",", "")

data["Vol."] = data["Vol."].str.replace("K","0")
data["Vol."] = data["Vol."].str.replace(".","")
data
```
